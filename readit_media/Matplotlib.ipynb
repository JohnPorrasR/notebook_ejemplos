# Uso de la librer√≠a Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y = x * 2
y
```

```python
plt.plot(x,y, 'red')
plt.title("Este es mi grafico")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

### Programaci√≥n orientada a objetos

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.8,0.8])
grafico.plot(x,y,'g')
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,'r')

grafico2 = figura.add_axes([0.2,0.4,0.3,0.3])
grafico2.plot(x,y,'b')
```

```python
figura, graficos = plt.subplots(nrows=2, ncols=2)

graficos[0][0].plot(x,y,'b')
graficos[0][1].plot(x,y,'r')
graficos[1][0].plot(x,y,'y')
graficos[1][1].plot(x,y,'g')

graficos[0][0].set_title('Azul')
graficos[0][1].set_title('Rojo')
graficos[1][0].set_title('Amarrillo')
graficos[1][1].set_title('Verde')
```

```python
# Tama√±o del grafico
figura = plt.figure(figsize=(10,2))
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,'r')
```

### Crear 2 graficos juntos

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x, x**2, label='Al cuadrado')
grafico.plot(x, x**3, label='Al cubo')
grafico.legend(loc=0)
```

### Linea del gr√°fico (grosor, transparencia y tipo)

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x, y, linewidth=10, alpha=.4, linestyle='-.')
```

### Marcadores en las intersecciones

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x, y, color='g', marker='o',markersize=5)
```
